```python
import tensorflow as tf
import numpy as np
import os
from tensorflow.examples.tutorials.mnist import input_data
```

```python


mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
Hidden_nodes = 100
Learning_rate = 0.1
Batch_size = 10000
Epochs = 100

x = tf.placeholder(tf.float32, [None, 784])
y = tf.placeholder(tf.float32, [None, 10])
w1 = tf.Variable(tf.random_normal([784, Hidden_nodes], stddev=(1/tf.sqrt(float(Hidden_nodes)))), name='W1')
b1 = tf.Variable(tf.random_normal([Hidden_nodes]), name='b1')
w2 = tf.Variable(tf.random_normal([Hidden_nodes, 10], stddev=(1/tf.sqrt(float(Hidden_nodes)))), name='W2')
b2 = tf.Variable(tf.random_normal([10]), name='b2')

z1 = tf.add(tf.matmul(x,w1),b1)
a1 = tf.nn.relu(z1)

z2 = tf.add(tf.matmul(a1,w2),b2)
a = tf.nn.softmax(z2)
a = tf.clip_by_value(a, 1e-10, 0.9999999)

cross_entropy = -tf.reduce_mean(tf.reduce_sum(y * tf.log(a) + (1 - y) * tf.log(1 - a), axis=1))
optimiser = tf.train.GradientDescentOptimizer(learning_rate=Learning_rate).minimize(cross_entropy)

init_op = tf.global_variables_initializer()
correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(a, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

saver = tf.train.Saver()
save_dir = 'checkpoints/'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)    
save_path = os.path.join(save_dir, 'test2')

with tf.Session() as sess:
    sess.run(init_op)
    
    total_batch = int(len(mnist.train.labels) / Batch_size)
    for Epoch in xrange(Epochs):
        for i in range(total_batch):
            
            batch_x, batch_y = mnist.train.next_batch(batch_size=Batch_size)
            _, c = sess.run([optimiser, cross_entropy], feed_dict={x: batch_x, y: batch_y}) 
            print "Cost at Batch {} is {}".format(i,c)
            
    print(sess.run(accuracy, feed_dict={x: mnist.test.images, y: mnist.test.labels}))
    print(sess.run(accuracy, feed_dict={x: mnist.validation.images, y: mnist.validation.labels}))
    saver.save(sess=sess, save_path=save_path)
    print("Model saved in path: %s" % save_path)
```
